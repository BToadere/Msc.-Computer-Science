```python
import itertools
import numpy as np


def escribir_linea(file, valor):

    with open(file, 'a') as fs:
        fs.writelines(valor + '\n')
    return

def proc_lineas(linea):

    linea_procesada=np.array(linea.strip().split(), dtype=np.float64)
    if np.any(linea_procesada < 0):
        escribir_linea('ascensos.txt', 'Error')
        return
    else:
        return linea_procesada

def pos_bajada_sostenida(linea):

    if linea is None:
        return
    
    # Objeto para almacenar las differrencias
    diff=np.diff(linea)
    tuplas_diff=zip(diff[:-1],diff[1:])
    iter_descartador = itertools.dropwhile(lambda tupla: np.any(np.array(tupla) >0), tuplas_diff)
    
    try:
        val_aux=next(iter_descartador)[0]
        indice = str(np.where(diff == val_aux)[0][0])
        escribir_linea('ascensos.txt', indice)

    except StopIteration:
        escribir_linea('ascensos.txt', 'No se encontraron valores negativos consecutivos.')

    return

with open('lanzamiento.txt', 'r') as file:
    lineas_vectorizadas= map(proc_lineas, file)
    
    list(map(pos_bajada_sostenida, lineas_vectorizadas))
```

```python
class compra():
    def __init__(self, linea):
        linea_limpia = linea.strip().split()
        self.dia = int(linea_limpia[0])
        self.hora = int(linea_limpia[1])
        self.minuto = int(linea_limpia[2])
        self.articulos = linea_limpia[3:-3]
        self.t.importe = float(linea_limpia[-2])
        self.t.tarjeta = int(linea_limpia[-1])

    def premio(self, gordo):
        if self.t.tarjeta == 0:
            return False
        elif any(map(lambda x: x[-4:] == gordo[-4:], self.articulos)):
            return 0.9*self.t.importe
        return False
    
gordo = '43458'#input('Introduzca el número del Gordo:')
presupuesto = 100 #float(input('Introduzca el presupuesto:'))

with open('diciembre.com', 'r') as file:
    for linea in file:
        c=compra(linea)
        premio = c.premio(gordo)
        if premio:
            premio = min(premio, presupuesto)
            presupuesto -= premio
            fs=open('sorteo.prem', 'a')
            fs.writelines(f'{c.t.tarjeta} {premio}\n')
            fs.close()
            
```

```python
import numpy as np
import re


class terminal_venta():
    def __init__(t):
        t.data={}

    def exec(t):

        while True:
            entrada = input('Tarjeta e importe: ')

            try:
                tar, imp = entrada.split()
            except:
                tar = entrada

            if not re.match(r'^\d+$', tar):
                break
            
            else:
                t.in_trajeta_importe(tar, imp)
                if t.comprobar_tarjeta():
                    t.actualizar()
                    print(f'Tarjeta acpetada. El importe es el {t.pct_importe_total()}% del máximo')
                else:
                    print('Tarjeta rechazada')

    def in_trajeta_importe(t, tarjeta, importe): 
        t.tarjeta = tarjeta
        t.importe = round(float(importe),2)

    def comprobar_tarjeta(t):
        tarjeta_aux = '#'+t.tarjeta
        try:
            suma_posiciones_pares = sum(map(lambda x: int(x), tarjeta_aux[-2:0:-2]))
            
            if suma_posiciones_pares % 6 == 0:
                return True
            else:
                return False
        except:
            return False

    def actualizar(t):

        if t.tarjeta in t.data.keys():
            importe_actual = t.data[t.tarjeta]
            t.data[t.tarjeta] = np.hstack((importe_actual,t.importe))
        else:
            t.data[t.tarjeta] = np.array(t.importe)

    def importe_total(t):
        return np.sum(t.data[t.tarjeta])
    
    def pct_importe_total(t):
        return round(100*t.importe/t.importe_total(),2)
    
    

if __name__ == '__main__':

    ter = terminal_venta()

    ter.exec()


```

```python
import math
import itertools
import functools

# Función que calcula el término k-ésimo de la serie de Fourier
def termino_serie(k, x):
    coef = 4 * (-1)**k / (k**2 * math.pi**2)
    return coef * math.cos(k * math.pi * x)

# Función que suma todos los términos de la serie hasta el término n usando programación funcional
def serie_fourier(x, n):
    # Constante inicial de la serie
    serie_inicial = 1 / 3
    
    # Generamos los términos de la serie desde k=1 hasta k=n
    terminos = map(lambda k: termino_serie(k, x), range(1, n + 1))
    
    # Sumamos todos los términos usando reduce
    suma_serie = functools.reduce(lambda acc, t: acc + t, terminos, serie_inicial)
    
    return suma_serie

# Función que calcula el valor de x^2
def cuadrado(x):
    return x**2

# Función que genera n puntos equidistantes en un intervalo [a, b]
def generar_puntos(a, b, num_puntos):
    return [a + (b - a) * i / (num_puntos - 1) for i in range(num_puntos)]

# Función que calcula el error medio entre la función x^2 y la serie de Fourier
def calcular_error_medio(a, b, num_puntos, n):
    puntos = generar_puntos(a, b, num_puntos)
    diferencias = [abs(cuadrado(x) - serie_fourier(x, n)) for x in puntos]
    return sum(diferencias) / len(diferencias)

# Función principal que encuentra el número mínimo de términos necesario para lograr el error deseado
def encontrar_num_terminos(a, b, num_puntos, limite_diferencia):
    for n in itertools.count(1):  # Genera un número creciente de términos
        error_medio = calcular_error_medio(a, b, num_puntos, n)
        if error_medio < limite_diferencia:
            return n, error_medio

# Entrada del usuario
a = -0.5#float(input("Introduce el extremo inferior del intervalo: "))
b = 0#float(input("Introduce el extremo superior del intervalo: "))
num_puntos = 5 #int(input("Introduce la cantidad de puntos a probar: "))
limite_diferencia = 0.05#float(input("Introduce el límite de diferencia media: "))

# Llamada a la función principal
terminos_necesarios, error_final = encontrar_num_terminos(a, b, num_puntos, limite_diferencia)

# Resultado final
print(f"Número de términos necesarios: {terminos_necesarios}")
print(f"Error medio alcanzado: {error_final}")
```

```python
import functools
import itertools

class Nodo:
    def __init__(self, capacidad):
        self.capacidad = capacidad  # Capacidad de mensajes que puede transmitir por segundo
        self.mensajes_en_espera = 0  # Mensajes que están esperando ser transmitidos

    def recibir_mensajes(self, cantidad):
        """
        Recibir mensajes y acumularlos en la cola de espera.
        """
        self.mensajes_en_espera += cantidad

    def transmitir_mensajes(self):
        """
        Transmitir mensajes de acuerdo a la capacidad del nodo.
        Devuelve los mensajes transmitidos y actualiza los mensajes en espera.
        """
        transmitidos = min(self.mensajes_en_espera, self.capacidad)
        self.mensajes_en_espera -= transmitidos
        return transmitidos

class RedDeTransmision:
    def __init__(self, capacidades, mensajes_iniciales):
        """
        Inicializa la red con una lista de capacidades para cada enlace y la cantidad de mensajes iniciales.
        """
        self.nodos = [Nodo(capacidad) for capacidad in capacidades]
        self.mensajes_iniciales = mensajes_iniciales

    def simular_transmision(self, segundos):
        """
        Simula la transmisión de mensajes durante un número de segundos.
        """
        # En el segundo 0, el nodo extremo recibe los mensajes iniciales
        self.nodos[0].recibir_mensajes(self.mensajes_iniciales)

        # Simulamos la transmisión para cada segundo
        for segundo in range(1, segundos + 1):
            print(f"Tiempo: {segundo}")
            mensajes_a_transmitir = 0

            # Iterar sobre cada nodo para transmitir mensajes
            for i, nodo in enumerate(self.nodos):
                # Los mensajes que el nodo tiene disponibles son los que transmite
                nodo.recibir_mensajes(mensajes_a_transmitir)
                mensajes_a_transmitir = nodo.transmitir_mensajes()
                print(f"Nodo {i}: {nodo.mensajes_en_espera} en espera, {mensajes_a_transmitir} transmitidos")

    def mensajes_en_nodo_en_tiempo(self, nodo_idx, segundo):
        """
        Devuelve la cantidad de mensajes que hay en un nodo específico en un segundo dado.
        """
        # Simula la transmisión hasta el segundo indicado
        self.simular_transmision(segundo)
        return self.nodos[nodo_idx].mensajes_en_espera

# Lectura de datos de entrada
cantidad_nodos = 4
mensajes_iniciales = 90000
# nombre_fichero = input("Ingrese el nombre del fichero con las capacidades de los enlaces: ")
capacidades = list((20000, 15000, 50000))
# Leer capacidades desde el archivo
# with open(nombre_fichero, 'r') as file:
#     capacidades = list(20000, 15000, 50000)

# Crear la red de transmisión
red = RedDeTransmision(capacidades, mensajes_iniciales)

# Pedir al usuario el segundo y el nodo a consultar
segundo_consulta = int(input("Ingrese el segundo a consultar: "))
nodo_consulta = int(input("Ingrese el nodo a consultar: "))

# Obtener el resultado y mostrarlo
mensajes_en_nodo = red.mensajes_en_nodo_en_tiempo(nodo_consulta, segundo_consulta)
print(f"En el segundo {segundo_consulta}, el nodo {nodo_consulta} tiene {mensajes_en_nodo} mensajes en espera.")
```
